# EVOLVE — YOLO Evaluation Notebook

This notebook is dedicated to the **evaluation and qualitative analysis** of
the YOLO-based object detection model trained on the EVOLVE dataset.

The objectives are:
- quantitative evaluation (mAP, precision, recall)
- qualitative inspection of detections
- identification of common failure modes

All dataset design choices, annotation rules, and scientific motivations
are documented in `evolve_workbook.qmd`.

```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import random
from pathlib import Path
```

## Load trained model

We load the **best-performing model weights** obtained during training.

```python
model = YOLO("../runs/detect/evolve_yolov8s/weights/best.pt")
```

## Quantitative evaluation

We evaluate the model using standard object detection metrics
on the validation or test split.

```python
metrics = model.val(
    data="../data/yolo/dataset.yaml",
    split="val"
)

metrics
```

## Qualitative inference on test images

We perform inference on a subset of test images in order to visually inspect:
- correct detections
- missed objects
- class confusions
- localization quality under extreme conditions

```python
model.predict(
    source="../data/yolo/images/test",
    conf=0.25,
    save=True,
    name="evolve_eval"
)
```

## Failure mode analysis

The following failure modes are of particular interest for the EVOLVE dataset:

- missed detections in very low-light conditions
- confusion between visually similar classes (e.g. `guitar` vs `micro`)
- coarse localization of collective patterns (`mosh_pit`, `hands_raised`)
- sensitivity to motion blur and partial occlusions

These cases are inspected manually using the prediction outputs
saved by YOLO.

## Notes and observations

- Detection performance varies strongly with illumination intensity.
- Small objects such as microphones are more frequently missed.
- Collective crowd patterns are detected with coarse but acceptable
  spatial localization.

These observations are discussed in the final project report.
