```python
from ultralytics import YOLO
import torch
```

## GPU availability check

```python
torch.cuda.is_available()
```

## Load pre-trained YOLO model

We start from a YOLOv8 model pre-trained on COCO and fine-tune it
on the EVOLVE dataset.

```python
model = YOLO("yolov8s.pt")
```

## Training configuration

Key parameters:
- image size: 640
- batch size: 16
- epochs: 50
- dataset format: YOLO

Training logs and weights are saved in the `runs/` directory.

```python
model.train(
    data="data/yolo/dataset.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    device=0,
    project="runs",
    name="evolve_yolo"
)
```

## Model export (optional)

The trained model can optionally be exported to other formats
(e.g. ONNX) for further analysis or deployment.

```python
model.export(format="onnx")
```
