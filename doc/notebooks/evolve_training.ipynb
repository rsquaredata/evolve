```python
from ultralytics import YOLO
import torch
```

## GPU availability check

```python
if torch.cuda.is_available():
    print("CUDA available")
    print(torch.cuda.get_device_name(0))
else:
    print("CUDA not available — training will be slow")
```

## Load pre-trained YOLO model

We start from a YOLOv8 model pre-trained on the COCO dataset
and fine-tune it on the EVOLVE dataset.

```python
model = YOLO("yolov8s.pt")
```

## Training configuration

Key parameters:
- image size: 640 (YOLO native resolution)
- batch size: 16
- epochs: 50
- dataset format: YOLO

Training logs, metrics, and model weights are automatically saved
in the `runs/` directory.

```python
model.train(
    data="../data/yolo/dataset.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    device="auto",
    name="evolve_yolov8s"
)
```

## Model export (optional)

> ⚠️ Model export is **optional** and **not required** for the course evaluation.
>
> This step is provided for completeness only and may be skipped.

```python
model.export(format="onnx")
```
